```python
#imports
import pandas as pd
```

```python
#read the cleaned dataset with a pandas dataframe
df = pd.read_csv('newdata.csv', parse_dates=['Date'])
```

```python
#sort the data by the values of the date and location column
df.sort_values(by=['Date','Location'], inplace=True)
```

```python
#have a first look on te data
df.head()
```

```python
#read the geospatial data that was brought from kaggle which is a dataset with the longitudes and latitudes of the 50 states.
dfgeov = pd.read_csv('statelatlong.csv')
```

```python
#sorting the geospatial data by the state column
dfgeov.sort_values(['State'], inplace= True)
```

```python
#have a first look on the geospatial data
dfgeov.head()
```

```python
#merge the geospatial data with the cleaned dataset by the values of both location and state values.
dfinal = df.merge(dfgeov, how='inner', left_on='Location', right_on='State')
```

```python
#having a look on the new data to check if the merge went right.
dfinal.head()
```

```python
#sort the data by the values of the data and location column
dfinal.sort_values(by=['Date','Location'], inplace=True)
```

```python
#drop the (Unnamed: 0) column.
dfinal.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#drop the (State) column because its values are duplicated.
dfinal.drop('State', axis=1, inplace=True)
```

```python
#have a final look on the data
dfinal.head()
```

```python
#checking if there are any messing values
dfinal.isnull().sum()
```

```python
#check the data type of the new data
dfinal.dtypes
```

```python
#upload the new data into csv file.
dfinal.to_csv('geodata.csv')
```
